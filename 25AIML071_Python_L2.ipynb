```python
x=int(input("Enter Integer Number:"))
if x%2==0:
    if x%3==0:
        print("Divide by both 2 and 3.")
    else:
        print("Divide by 2 and not by 3.")
elif x%3==0:
    print("Divide by 3 and not by 2.")
else:
    print("Not divide by 2 and 3.")

        
```

Branching Programs

```python
x= int(input("Enter First Number:"))
y= int(input("Enter Second Number:"))
z=int(input("Enter the third number:"))
if x<y and x<z:
    print(f"{x} is minimum.")
elif y<z:
    print(f"{y} is minimum.")
else:
    print(f"{z} is minimum.")
```
